### Matplotlib Subplot


```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 5)
y = x ** 3
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax[0, 1].plot([1,2,3,4],[1,4,9,16],'go')
ax[1, 0].plot(x,y,'r^')
ax[0, 1].set_title('Squares')
ax[1, 0].set_title('Cubes')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
division_average_marks = [70, 82, 73, 65, 68]
variance = [5, 8, 7, 6, 4]

plt.barh(divisions, division_average_marks, xerr=variance, color='g')
plt.title('Bar graph')
plt.xlabel('Mark')
plt.ylabel('Divisions')

plt.show()
```

 to cc biu  thanh xp chng theo chiu ngang, ta s dng phng thc `bar()` hai ln v chuyn cc i s trong  ta  cp n `index` v `width` ca biu  thanh  xp chng theo chiu ngang. Ngoi ra, ch  vic s dng hai phng thc `legend()` c s dng  hin th ch gii ca biu  v `xticks()`  gn nhn trc `x` da trn v tr ca cc thanh.

```python
import numpy as np
import matplotlib.pyplot as plt

divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 61, 70]

index = np.arange(5)
width = 0.3

plt.bar(index, division_average_marks, width, color='g', label='Division Marks')
plt.bar(index+width, boys_average_marks, width, color='b', label='Boys Marks')

plt.title('Horizontally stacked bar graphs')

plt.xlabel('Divisions')
plt.ylabel('Marks')
plt.xticks(index + width/2, divisions)

plt.legend(loc='best')

plt.show()
```

Tng t,  xp theo chiu dc cc biu  thanh vi nhau, chng ta c th s dng i s `bottom` v  cp n biu  thanh m chng ta mun xp chng bn di lm gi tr ca n.

```python
import numpy as np
import matplotlib.pyplot as plt

divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
boys_average_marks = [68, 67, 77, 61, 70]
girls_average_marks = [72, 97, 69, 69, 66]

index = np.arange(5)
width = 0.3

plt.bar(index, boys_average_marks, width, color='g', label='Boys Marks')
plt.bar(index, girls_average_marks, width, color='b', label='Girls Marks', bottom=boys_average_marks)

plt.title('Vertically stacked bar graphs')

plt.ylabel('Marks')
plt.xlabel('Divisions')
plt.xticks(index + width/2, divisions)

plt.legend(loc='best')

plt.show()
```
